### Cleaning the Dataset

```python
## checking for duplicates

df.duplicated().sum()
```

```python
## dealing with outliers

q1 = df["Rating"].quantile(0.25)
q3 = df["Rating"].quantile(0.75)
```

```python
iqr = q3 - q1
```

```python
upper_limit = q3 + (1.5*iqr)
lower_limit = q3 - (1.5*iqr)

print(lower_limit, upper_limit)
```

```python
df.loc[(df["Rating"] > upper_limit) | (df["Rating"] < lower_limit)]
```

```python
## Checking for outliers in the gross income column

q1 = df["gross income"].quantile(0.25)
q3 = df["gross income"].quantile(0.75)

iqr = q3 - q1

upper_limit = q3 + (1.5*iqr)
lower_limit = q3 - (1.5*iqr)

print(lower_limit, upper_limit)
```

```python
df.loc[(df["gross income"] > upper_limit) | (df["gross income"] < lower_limit)]
```

```python
df1 = df.loc[(df["gross income"] < upper_limit) & (df["gross income"] > lower_limit)]
df1
```

```python
print("before removing outliers:",len(df))
print("after removing outliers:", len(df1))
print("outliers:", len(df) - len(df1))
```

```python
sns.boxplot(df1["gross income"])
```

### Building Linear Regression Model

```python
sales1 = sales.drop(["Invoice ID", "Branch", "City", 'Customer type', "Date","Time", 'Gender', 'Product line', "Payment"], axis = 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = sales1.drop(["Rating"], axis = 1)
y = sales1["Rating"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr.predict(X_test)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, pred)
```

#### Plotting the predicted values

```python
plt.figure(figsize = (10,10))
plt.scatter(pred, y_test, s = 250, color = "green")
plt.xlabel("Predicted Value")
plt.ylabel("Ratings")
plt.grid()
plt.title("Predicted Value vs Ratings")
plt.show()
```
